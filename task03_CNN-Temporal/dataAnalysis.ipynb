```python
import os
import json

# 최상위 경로 설정
root_dir = '../data/CAT_image'
json_count = 0


# 하위 폴더 개수 세기
folder_count = sum(
    os.path.isdir(os.path.join(root_dir, name))
    for name in os.listdir(root_dir)
)

print(f"📁 폴더 개수: {folder_count}개")

for folder_name in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder_name)

    if os.path.isdir(folder_path):
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.json'):
                json_count += 1

print(f"📦 JSON 파일 개수: {json_count}개")
```

```python
# df = pd.read_csv("../data/normalized_meta_info.csv")
df = pd.read_csv("../data/meta_info.csv")
df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# situation 라벨별 샘플 수 확인
label_counts = df['situation'].value_counts()

# 5개 이상인 클래스만 필터링
valid_labels = label_counts[label_counts >= 50].index
df_filtered = df[df['situation'].isin(valid_labels)]


set(df_filtered['situation'].to_list())
```

```python
df_filtered['action'] = df_filtered['action'].apply(lambda x: action_label_map.get(x, x))
```

```python
label_columns = ['action', 'emotion', 'situation']

for col in label_columns:
     print(f"🔹 {col} 분포:")
     print(df_filtered[col].value_counts())
     unique_labels = sorted(df_filtered[col].unique())
     print(f"🔹 {col} 라벨 셋 ({len(unique_labels)}개):")
     for label in unique_labels:
         print(f"  - {label}")
     print()
     print()
```

```python
re_df_filtered = df_filtered[df_filtered['frames']>=80]
```

```python
label_columns = ['action', 'emotion', 'situation']

for col in label_columns:
     print(f"🔹 {col} 분포:")
     print(re_df_filtered[col].value_counts())
     unique_labels = sorted(re_df_filtered[col].unique())
     print(f"🔹 {col} 라벨 셋 ({len(unique_labels)}개):")
     for label in unique_labels:
         print(f"  - {label}")
     print()
     print()
```

```python
# 필터된 데이터로 stratify
# 예: 'situation' 기준 stratify하여 train/val/test 분할
train_val, test = train_test_split(re_df_filtered, test_size=0.2, random_state=42, stratify=re_df_filtered['situation'])
train, val = train_test_split(train_val, test_size=0.25, random_state=42, stratify=train_val['situation'])  # 0.25*0.8=0.2

print(f"Train size: {len(train)}, Val size: {len(val)}, Test size: {len(test)}")
print("Train action distribution:\n", train['situation'].value_counts(normalize=True))
print("Val action distribution:\n", val['situation'].value_counts(normalize=True))
print("Test action distribution:\n", test['situation'].value_counts(normalize=True))
```

```python
train, val, test
train.to_csv("train_2nd.csv", index=False)
val.to_csv("val_2nd.csv", index=False)
test.to_csv("test_2nd.csv", index=False)
```
